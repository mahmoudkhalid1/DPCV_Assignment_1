```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
Original_img = np.zeros([512, 512])
Original_img[0:256,0 :256] = 40
Original_img[256:512,0 :256] = 80 
Original_img[256:512,0 :256] = 80 
Original_img[256:512,256 :512] = 220 
plt.imshow(Original_img, cmap='gray')
plt.imsave('OriginalImage_DS3.png', Original_img, cmap='gray')
```

```python
def add_noise(img, x1,y1,x2,y2):
 
    # Getting the dimensions of the image
    row , col = img.shape
     
    # Randomly pick some pixels in the
    # image for coloring them white
    # Pick a random number between x1 and y1
    number_of_pixels = random.randint(x1, y1)
    for i in range(number_of_pixels):
       
        # Pick a random y coordinate
        y_coord=random.randint(0, row - 1)
         
        # Pick a random x coordinate
        x_coord=random.randint(0, col - 1)
         
        # Color that pixel to white
        img[y_coord][x_coord] = 255
         
    # Randomly pick some pixels in
    # the image for coloring them black
    # Pick a random number between x2 and y2
    number_of_pixels = random.randint(x2 , y2)
    for i in range(number_of_pixels):
       
        # Pick a random y coordinate
        y_coord=random.randint(0, row - 1)
         
        # Pick a random x coordinate
        x_coord=random.randint(0, col - 1)
         
        # Color that pixel to black
        img[y_coord][x_coord] = 0
         
    return img
 
```

```python
temp=Original_img
Light_noise = add_noise(temp,0,450,0,450)
plt.imsave('LightNoise_DS3.png', Light_noise, cmap='gray')
plt.imshow(Light_noise,cmap='gray')
```

```python
Meduim_noise =add_noise(temp,300,900,300,900)
plt.imshow(Meduim_noise,cmap='gray')
plt.imsave('MeduimNoise_DS3.png', Meduim_noise, cmap='gray')

```

```python
Hight_noise=add_noise(temp,1500,3500,1500,3500)
plt.imshow(Hight_noise,cmap='gray')
plt.imsave('HighNoise_DS3.png', Hight_noise, cmap='gray')

```

```python
def QuantizationGrey(img,IQ):
    n = img.shape[0]
    n2= img.shape[1]
    for i in range (0,n):
        for j in range (0,n2):
            newVal=(np.floor((img[i][j]/IQ))*IQ)
            img[i][j]=newVal
    return img

def QuantizationColor(img,IQ):
    n = img.shape[0]
    n2= img.shape[1]
    for i in range (0,n):
        for j in range (0,n2):
            newValR=(np.floor((img[i][j][0]/IQ))*IQ)
            newValG=(np.floor((img[i][j][1]/IQ))*IQ)
            newValB=(np.floor((img[i][j][2]/IQ))*IQ)
            img[i][j][0]=newValR
            img[i][j][1]=newValG
            img[i][j][2]=newValB


    return img
```

```python
imgPlaneG= plt.imread('planeGray.jpeg')
plt.imshow(imgPlaneG, cmap='gray')
```

```python
plt.imshow(QuantizationGrey(imgPlaneG,16), cmap='gray')
```

```python
plt.imshow(QuantizationGrey(imgGreensG,16), cmap='gray')
```
